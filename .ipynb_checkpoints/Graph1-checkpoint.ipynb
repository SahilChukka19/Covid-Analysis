#                                      UK Death Rate by Vaccination Status

```python
#Imports
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
plt.style.use('ggplot')
```

```python
#Path to the dataset
data_path = 'UKcovid.xlsx'
df = pd.read_excel(data_path)
```

```python
#Checking the dataset,whether it contains any null values or duplicates
df.shape
```

```python
df.head()
```

```python
df.info
```

```python
#pivoting the dataset
df1=df.pivot(index='Age group', columns='Vaccination status', values = 'Deaths involving COVID-19')
```

```python
#horizonatal bar graph 
df1.plot(kind='barh', stacked=False, figsize=(12, 8))
plt.xlabel('Age Group')
plt.ylabel('Deaths involving COVID-19')
plt.title('Vaccination Status by Age Group')
plt.show()
```

```python
df2 = df.pivot(index='Age group', columns='Vaccination status', values = 'Non-COVID-19 deaths')
```

```python
df2.plot(kind='barh', stacked=False, figsize=(12, 8))
plt.xlabel('Age Group')
plt.ylabel('Non-COVID-19 deaths')
plt.title('Vaccination Status by Age Group')
plt.show()
```

```python
df2 = df.pivot(index='Age group', columns='Vaccination status', values = 'All deaths')
```

```python
df2.plot(kind='barh', stacked=False, figsize=(12, 8))
plt.xlabel('Age Group')
plt.ylabel('All Deaths')
plt.title('Vaccination Status by Age Group')
plt.show()
```
