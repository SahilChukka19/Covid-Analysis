# COVID 19 IN INDIA

```python
#Imports
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
```

```python
#datapath
df = pd.read_excel('districts.xlsx')
```

```python
#Checking for null values, duplicates and trying to get a better understanding of the strutcure of the data
df.head()
```

```python
df.duplicated()
```

```python
df.info
```

```python
# visualization using a Sorted Bar graph
bar_data = df[df['Date'] == '2020-04-26'].groupby('State')['Confirmed'].sum().reset_index()
bar_data = bar_data.sort_values(by='Confirmed')
plt.figure(figsize=(15, 6))
plt.bar(bar_data['State'], bar_data['Confirmed'], color='red')
plt.xlabel('State')
plt.ylabel('Confirmed Cases')
plt.title('Total Confirmed Cases by State on 2020-04-26')
plt.xticks(rotation=90)
plt.show()
```

```python
# Visualization of the cases trend in Maharashtra using Line graph
line_data = df[df['State'] == 'Maharashtra'].groupby('Date')['Confirmed'].sum().reset_index()
plt.figure(figsize=(15, 6))
plt.plot(line_data['Date'], line_data['Confirmed'], marker='o', color='orange')
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')
plt.title('Total Confirmed Cases Trend in Maharashtra')
plt.xticks(rotation=90)
plt.show()
```

```python
# Visualization of the Confirmed cases in Andhra Pradesh using Pie chart on 2020-04-26
pie_data = df[(df['State'] == 'Andhra Pradesh') & (df['Date'] == '2020-04-26')]
plt.figure(figsize=(8,8))
plt.pie(pie_data['Confirmed'], labels=pie_data['District'], autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Distribution of Confirmed Cases in Andhra Pradesh on 2020-04-26')
plt.show()
```
